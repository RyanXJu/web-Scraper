### choose only condo

```python
condo = data.loc[data.title == "Condo for sale", :]
condo.shape
```

```python
#### these features might be very important
condo["Net area"].isnull().sum() 
condo["bedrooms"].isnull().sum()
condo["rooms"].isnull().sum()
```

```python
condo.isnull().sum(axis=0) <= condo.shape[0]*0.1
```

```python
# remove columns with more than 10% missing values
condo = condo.loc[:, (condo.isnull().sum(axis=0) <= condo.shape[0]*0.1)]
condo.head(5)
```

```python
condo.columns
```

```python
# drop address
condo.drop(columns = ["title", "address"])
```

```python
# normalize coordinate
# Montreal lat = 45.7, long = -73.5673 (by google) 
# 45.703282, -73.476106  to 45.356596, -73.991630
lat_mtl_min = 45
lat_mtl_max = 46
long_mtl_min = -74
long_mtl_max = -73
condo.lat = condo.lat-lat_mtl_min
condo.long = condo.long-long_mtl_min
```

```python
envir = ['Restaurants', 'Shopping', 'Vibrant', 'Cycling friendly', 'Car friendly','Historic', 'Quiet',
         'Elementary Schools', 'High Schools', 'Parks', 'Nightlife', 'Groceries', 
         'Daycares', 'Pedestrian friendly', 'Cafes', 'Transit friendly','Greenery',]
condo.loc[ :, envir] = condo.loc[ :, envir] /10
```

```python
condo[envir] = condo[envir].fillna(0)
condo[envir].isnull().any() 
```

```python
# transform type to binary
condo["Condominium type"] = condo["Condominium type"] == "Divided"
```

```python
condo["Condominium type"] = condo["Condominium type"].astype(int)
condo["Condominium type"].value_counts()
```

```python
# year build
condo.loc[(condo["Year built"]=="To be built, New") | (condo["Year built"]== "Under construction, New"), "Year built"] = "2020"
condo["Year built"] = condo["Year built"].str.extract("(\d\d\d\d)")

# impute unkown age to 1900
condo["Year built"] = condo["Year built"].fillna(1900)
condo[condo["Year built"].isnull()].index.tolist()

condo["Year built"] = condo["Year built"].astype(int)
```

```python
condo["Year built"].describe()
condo["Year built"] = (condo["Year built"] - 1800)/(2020-1800)
```

```python
condo["Year built"].describe()
```

```python
### rooms, impute nan to 2
condo.rooms = condo.rooms.fillna(2)
```

```python
### other statistics
stats = ['Unemployment rate 2016 ',
       'Less than $50,000 (%)', 'Between $50,000 and $80,000 (%)',
       'Between $80,000 and $100,000 (%)', 'Between $100,000 and $150,000 (%)',
       'More than $150,000 (%)', '1-person households (%)',
       '2-person households (%)', '3-person households (%)',
       '4-person households (%)', '5-person or more households (%)',
       'Couples without children at home (%)',
       'Couples with children at home (%)', 'Single-parent families (%)',
       'Owners (%)', 'Renters (%)', 'Before 1960 (%)',
       'Between 1961 and 1980 (%)', 'Between 1981 and 1990 (%)',
       'Between 1991 and 2000 (%)', 'Between 2001 and 2010 (%)',
       'Between 2011 and 2016 (%)', 'Single-family homes (%)',
       'Semi-detached or row houses (%)',
       'Buildings with less than 5 floors (%)',
       'Buildings with 5 or more floors (%)', 'Mobile homes (%)',
       'University (%)', 'College (%)', 'Secondary (high) school (%)',
       'Apprentice or trade school diploma (%)', 'No diploma (%)',
       'Non-immigrant population (%)', 'Immigrant population (%)',
       'French (%)', 'English (%)', 'Others languages (%)', 'walk_score']
condo.loc[:,stats] = condo.loc[:,stats].fillna(0)
condo.loc[:,stats] = condo.loc[:,stats]/100
```

```python
condo[stats].isnull().any() 
```

```python
### population stats 
#'Population 2016 ', 
#'Population variation between 2011 2016 ',
#'Population density '
condo["Population 2016 "].describe()
popMax = condo["Population 2016 "].max()
popMin = condo["Population 2016 "].min()
condo["Population 2016 "] = (condo["Population 2016 "] - popMin)/(popMax-popMin)
```

```python
condo["Population variation between 2011 2016 "].describe()
condo["Population variation between 2011 2016 "] = condo["Population variation between 2011 2016 "]/9
```

```python
condo["Population density "].describe()
popDenMax = condo["Population density "].max()
popDenMin = condo["Population density "].min()
condo["Population density "] = (condo["Population density "] - popDenMin)/(popDenMax-popDenMin)
```

```python
condo.describe()
```

```python
import torch
print(torch.__version__)
### pytorch 1.60 will have error while transfer model or tensor to gpu
### RuntimeError: CUDA error: an illegal memory access was encountered

from torch.autograd import Variable
import torch.nn.functional as F
import torch.utils.data as Data
import torch.nn as nn
import torch.optim as optim
from sklearn.preprocessing import StandardScaler 
from torch.utils.data import Dataset, DataLoader

import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
import imageio
import sklearn

torch.manual_seed(42)    # reproducible
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
EPOCHS = 200
BATCH_SIZE = 64
```

```python
y_train = torch.from_numpy(np.array(y_train).astype(np.float32))
torch.FloatTensor(y_train)
```

```python
## train data
class trainData(Dataset):
    
    def __init__(self, X_data, y_data):
        self.X_data = X_data
        self.y_data = y_data
        
    def __getitem__(self, index):
        return self.X_data[index], self.y_data[index]
        
    def __len__ (self):
        return len(self.X_data)


train_data = trainData(torch.FloatTensor(X_train), torch.FloatTensor(y_train))


## test data    
class testData(Dataset):
    
    def __init__(self, X_data):
        self.X_data = X_data
        
    def __getitem__(self, index):
        return self.X_data[index]
        
    def __len__ (self):
        return len(self.X_data)
    

test_data = testData(torch.FloatTensor(X_test))
```

```python
train_loader = DataLoader(dataset=train_data, batch_size=BATCH_SIZE, shuffle=True)
test_loader = DataLoader(dataset=test_data, batch_size=1)
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
print(device)
```

```python
# define a network
net = torch.nn.Sequential(
        torch.nn.Linear(63, 200),
        torch.nn.LeakyReLU(),
        torch.nn.Linear(200, 100),
        torch.nn.LeakyReLU(),
        torch.nn.Linear(100, 1),
    )
net.to(device)
```

```python
optimizer = torch.optim.Adam(net.parameters(), lr=0.001)
loss_func = torch.nn.MSELoss()  # this is for regression mean squared loss
```

```python
net.train()
for e in range(1, EPOCHS+1):
    epoch_loss = 0
    epoch_acc = 0
    for X_batch, y_batch in train_loader:
        X_batch, y_batch = X_batch.to(device), y_batch.to(device)
        optimizer.zero_grad()
        
        y_pred = net(X_batch)
        
        loss = loss_func(y_pred, y_batch.unsqueeze(1))
        
        loss.backward()
        optimizer.step()
        
        epoch_loss += loss.item()
        

    print(f'Epoch {e+0:03}: | Loss: {epoch_loss/len(train_loader):.5f}')
```

```python
y_pred_list = []

net.eval()
with torch.no_grad():
    for X_batch in test_loader:
        X_batch = X_batch.to(device)
        y_test_pred = net(X_batch)
#         print(y_test_pred)

        y_pred_list.append(y_test_pred.cpu().numpy())
        
y_pred_list = [a.squeeze().tolist() for a in y_pred_list]
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_pred_list)
```
